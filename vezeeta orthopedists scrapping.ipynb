# Importing Libraries

```python
# importing libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
# creating placeholders for data
titles = []
doctor_names = []
speciality = []
sub_speciality = []
address = []
price = []
rating = []
views = []
urls = []
```

# Getting doctors URL

```python
# loop for pages to get urls and some information
for i in range(1, 125):

    # Generating a request for vezeeta.com
    url = f'https://www.vezeeta.com/en/doctor/orthopedics/egypt?page={i}'
    response = requests.get(url)

    # Getting HTML code
    html = response.text

    # Let's parse this HTML code
    soup = BeautifulSoup(html, 'html.parser')

    # Doctor titles
    title_elements = soup.find_all('span',
         {'class': 'DoctorCardSubComponentsstyle__Text-sc-1vq3h7c-14 DoctorCardSubComponentsstyle__TitleText-sc-1vq3h7c-16 gqVrId hEuQjC'})
    for title_element in title_elements:
        titles.append(title_element.text)
    print('title',i)
    
    # Doctor name
    doctor_name = soup.find_all('h4',
         {'class': 'DoctorCardSubComponentsstyle__Text-sc-1vq3h7c-14 DoctorCardSubComponentsstyle__DoctorNameText-sc-1vq3h7c-15 gqVrId Wpycp'})
    for doctor in doctor_name:
        doctor_names.append(doctor.text)
    print('name',i)

    # Doctor speciality
    speciality_element = soup.find_all('p', 
            {'class': 'DoctorCardSubComponentsstyle__Text-sc-1vq3h7c-14 DoctorCardSubComponentsstyle__DescText-sc-1vq3h7c-17 gqVrId esZVig'})
    for specialit in speciality_element:
        speciality.append(specialit.text)
    print('speciality', i)

    # Doctor urls
    link_elements = soup.find_all('a', {'class': 'CommonStylesstyle__TransparentA-sc-1vkcu2o-2 cTFrlk'})
    for link_element in link_elements:
        urls.append('https://www.vezeeta.com' + link_element.get('href'))
    print('url',i)
```

```python
print(len(urls), len(titles), len(doctor_names),len(speciality))
```

```python
# splitting urls 
url1 = urls[:200]
url2 = urls[200:400]
url3 = urls[400:600]
url4 = urls[600:800]
url5 = urls[800:1000]
url6 = urls[1000:]
```

# Scraping the data

```python
#Data Scrapping function definition
def scrap(doctor_url):
    count = 0
    #doctors urls loop
    for url in doctor_url:
        count += 1
        response = requests.get(url)
        html = response.text
        soup = BeautifulSoup(html, 'html.parser')

        # Doctor sub-speciality
        sub_speciality_elements = soup.find_all('a', {'class': 'ProfileDoctorCardstyle__Link-sc-1u0n0d5-7 dQQSMn'})
        sub_specialities = [element.text for element in sub_speciality_elements]
        if sub_specialities:
            sub_speciality.append(sub_specialities)
        else:
            sub_speciality.append('np.nan')
        print('sub_speciality', count)
    
        # Doctor views
        view_elements = soup.find_all('span', {'class': 'ProfileDoctorCardstyle__ViewCount-sc-1u0n0d5-26 bOkJak'})
        views.append([element.text for element in view_elements])
        print('views', count)
    
        # Doctor price
        price_elements = soup.find_all('span', {'class': 'ReservationDetailsstyle__ItemText-sc-iq98nc-18 ehzOZm'})
        prices = [element.text for element in price_elements]
        if prices:
            price.append(prices)
        else:
            price.append('np.nan')
        print('price', count)
        
        # Doctor Rating
        rating_elements = soup.find_all('span', {'class': 'DoctorReviewsstyle__RatingText-sc-o1mk5u-14 gRAVjE'})
        ratings = [element.text for element in rating_elements]
        if ratings:
            rating.append(ratings)
        else:
            rating.append('np.nan')
        print('rating', count)

        # Doctor address
        address_element = soup.find('span', {'itemprop': 'address'})
        if address_element is not None:
            address.append(address_element.text)
        else:
            address.append(np.nan)
        print('address', count)
                
```

```python
# real_time scrapping with sleep for 1 minute
scrap(url1)
time.sleep(20)
scrap(url2)
time.sleep(20)
scrap(url3)
time.sleep(20)
scrap(url4)
time.sleep(20)
scrap(url5)
time.sleep(20)
scrap(url6)
```

# Cleaning the data

```python
df = df.replace('not_specified',np.nan)
```

### Price column

```python
df['Price'] = df['Price'].astype(str)
df['Price'] = pd.to_numeric(df['Price'].str.replace('[^\d.]', '', regex=True), errors='coerce')
```

```python
df['Price'] = df['Price'].astype('Int64')  # Convert to integer type
```

### Doctor_Rating column

```python
df['Doctor_Rating'] = df['Doctor_Rating'].astype(str)
df['Doctor_Rating'] = pd.to_numeric(df['Doctor_Rating'].str.replace('[^\d.]', '', regex=True), errors='coerce')
```

```python
df['Doctor_Rating'] = df['Doctor_Rating'].astype('float')  # Convert to integer type
```

### Views column

```python
df['Views'] = df['Views'].astype(str)
df['Views'] = pd.to_numeric(df['Views'].str.replace('[^\d.]', '', regex=True), errors='coerce')
```

```python
df['Views'] = df['Views'].astype('Int64')  # Convert to integer type
```

### Sub_speciality column

```python
df['Sub_speciality'] = df['Sub_speciality'].astype(str)
df['Sub_speciality'] = df['Sub_speciality'].str.strip().str.replace(r'\[|\]', '', regex=True)
df['Sub_speciality'] = df['Sub_speciality'].str.replace("'", "")
```
